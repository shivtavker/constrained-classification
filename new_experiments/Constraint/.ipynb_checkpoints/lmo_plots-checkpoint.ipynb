```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
```

```python
plt.rcParams.update({'font.size': 14})
```

```python
dataset = "abalone"
split = "test"
```

```python
obj_sbfw = np.load("./lmo-results/sbfw_" + dataset + "_" + split + "_scores.npy", allow_pickle=True)
obj_3pplug = np.load("./lmo-results/3pplug_" + dataset + "_" + split + "_scores.npy", allow_pickle=True)
obj_el = np.load("./lmo-results/elp_" + dataset + "_" + split + "_scores.npy", allow_pickle=True)
```

```python
cons_sbfw = np.load("./lmo-results/sbfw_" + dataset + "_" + split + "_cons.npy", allow_pickle=True)
cons_3pplug = np.load("./lmo-results/3pplug_" + dataset + "_" + split + "_cons.npy", allow_pickle=True)
cons_el = np.load("./lmo-results/elp_" + dataset + "_" + split + "_cons.npy", allow_pickle=True)
```

```python
obj_sbfw_new = np.load("./lmo-results/sbfw_new_" + dataset + "_" + split + "_scores.npy", allow_pickle=True)
obj_3pplug_new = np.load("./lmo-results/3pplug_new_" + dataset + "_" + split + "_scores.npy", allow_pickle=True)

cons_sbfw_new = np.load("./lmo-results/sbfw_new_" + dataset + "_" + split + "_cons.npy", allow_pickle=True)
cons_3pplug_new = np.load("./lmo-results/3pplug_new_" + dataset + "_" + split + "_cons.npy", allow_pickle=True)
```

```python
mean_score_sbfw = 1 - np.median(obj_sbfw, axis=1)
mean_score_3pplug = 1 - np.median(obj_3pplug, axis=1)
mean_score_el = 1 - np.median(obj_el, axis=1)
```

```python
mean_cons_sbfw = np.median(cons_sbfw, axis=1)
mean_cons_3pplug = np.median(cons_3pplug, axis=1)
mean_cons_el = np.median(cons_el, axis=1)
```

```python
T_vals = [10, 50, 75, 100, 250, 500, 1000, 2000, 3000, 4000, 5000]
```

```python
fig = plt.figure(figsize=(5,3))
ax = fig.add_axes([0, 0, 1, 1])

plt.plot(T_vals, mean_score_sbfw[:len(T_vals)], label="SplitFW", linestyle="-.")
plt.plot(T_vals, mean_score_3pplug[:len(T_vals)], label="ConGDA", linestyle=":")
plt.plot(T_vals, mean_score_el[:len(T_vals)], label="ConElp")

# ax.yaxis.set_major_formatter(mtick.FormatStrFormatter('%.2e'))

# plt.ylim(0.15)
plt.ylabel("H-mean Loss")
plt.xlabel("No. LMO Calls")
plt.legend(fontsize=14)
plt.savefig("./lmo-plots/" + dataset + "_" + split + "_score.pdf", bbox_inches="tight")
plt.show()
```

```python
fig = plt.figure(figsize=(5,3))
ax = fig.add_axes([0, 0, 1, 1])

plt.plot(T_vals, mean_cons_sbfw[:len(T_vals)], label="SplitFW", linestyle="-.")
plt.plot(T_vals, mean_cons_3pplug[:len(T_vals)], label="ConGDA", linestyle=":")
plt.plot(T_vals, mean_cons_el[:len(T_vals)], label="ConElp")

ax.fill_between([0, T_vals[-1]*1.05], 0, 0.01, facecolor="green", alpha=0.25)

plt.ylabel("Coverage Value")
plt.xlabel("No. LMO Calls")
plt.ylim(0.0, 0.12)
plt.xlim(0, T_vals[-1]*1.05)
plt.legend(fontsize=14)
plt.savefig("./lmo-plots/" + dataset + "_" + split + "_cons.pdf", bbox_inches="tight")
plt.show()
```
