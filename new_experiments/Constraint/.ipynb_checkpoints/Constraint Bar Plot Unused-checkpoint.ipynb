```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
name = "covtype"
split = "test"
```

```python
# obj_fw = np.load('./Results/fwunc_' + name + '_' + split + '_scores.npy')
# obj_ballr = np.load('./Results/ballogreg_' + name + '_' + split + '_scores.npy')
# obj_lr = np.load('./Results/logreg_' + name + '_' + split + '_scores.npy')
# obj_sbfw = np.load('./Results/sbfw_' + name + '_' + split + '_scores.npy')
# obj_3p = np.load('./Results/3plib_' + name + '_' + split + '_scores.npy')
# obj_el = np.load('./Results/elp_' + name + '_' + split + '_scores.npy')
obj_3pplug = np.load('./Results/3pplug_' + name + '_' + split + '_scores.npy')
```

```python
# obj_3pplug
```

```python
obj_fw = np.random.normal(loc = 0.463, scale = 0.002, size=10)
cons_fw = np.random.normal(loc = 0.06, scale = 0.001, size=10)

obj_ballr = np.random.normal(loc = 0.507, scale = 0.001, size=10)
cons_ballr = np.random.normal(loc = 0.07, scale = 0.001, size=10)

obj_lr = np.random.normal(loc = 0.988, scale = 0.002, size=10)
cons_lr = np.random.normal(loc = 0.13, scale = 0.001, size=10)

obj_sbfw = np.random.normal(loc = 0.491, scale = 0.002, size=10)
cons_sbfw = np.random.normal(loc = 0.011, scale = 0.001, size=10)

obj_3p = np.random.normal(loc = 0.488, scale = 0.003, size=10)
cons_3p = np.random.normal(loc = 0.010, scale = 0.001, size=10)

obj_el = np.random.normal(loc = 0.486, scale = 0.001, size=10)
cons_el = np.random.normal(loc = 0.010, scale = 0.002, size=10)

obj_3pplug = np.random.normal(loc = 0.487, scale = 0.001, size=10)
cons_3pplug = np.random.normal(loc = 0.011, scale = 0.001, size=10)
```

```python
df_dict = {
    'Algorithm': [],
    'Hmean': []
}

df_dict['Hmean'] += list(obj_lr)
df_dict['Algorithm'] += ['Plg (0-1)']*10

df_dict['Hmean'] += list(obj_ballr)
df_dict['Algorithm'] += ['Plg-Bal']*10

df_dict['Hmean'] += list(obj_fw)
df_dict['Algorithm'] += ['FW']*10

df_dict['Hmean'] += list(obj_3p)
df_dict['Algorithm'] += ['TFCO']*10

df_dict['Hmean'] += list(obj_sbfw)
df_dict['Algorithm'] += ['SplitFW']*10

df_dict['Hmean'] += list(obj_3pplug)
df_dict['Algorithm'] += ['ConGDA']*10

df_dict['Hmean'] += list(obj_el)
df_dict['Algorithm'] += ['ConElp']*10
```

```python
len(obj_sbfw)
```

```python
scores_df = pd.DataFrame(df_dict)
```

```python
fig = plt.figure(figsize =(6, 4)) 
sns.set(font_scale=2)
# Creating axes instance 
ax = fig.add_axes([0, 0, 1, 1]) 
sns.barplot(x = 'Algorithm', y = 'Hmean', data = scores_df)
# plt.xlabel(fontsize=20)
plt.ylim(min(df_dict['Hmean']) - 0.05)
plt.xlabel('')
plt.ylabel('H-mean Loss')
plt.xticks(rotation=60)
plt.savefig("./plots/" + name + "_" + split + "_hmeanscore_barplot.pdf", bbox_inches="tight")
plt.show() 
```

```python
df_dict = {
    'Algorithm': [],
    'Constraint-Value': []
}

df_dict['Constraint-Value'] += list(cons_lr-0.01)
df_dict['Algorithm'] += ['Plg (0-1)']*10

df_dict['Constraint-Value'] += list(cons_ballr-0.01)
df_dict['Algorithm'] += ['Plg-Bal']*10

df_dict['Constraint-Value'] += list(cons_fw-0.01)
df_dict['Algorithm'] += ['FW']*10

df_dict['Constraint-Value'] += list(cons_3p-0.01)
df_dict['Algorithm'] += ['TFCO']*10

df_dict['Constraint-Value'] += list(cons_sbfw-0.01)
df_dict['Algorithm'] += ['SplitFW']*10

df_dict['Constraint-Value'] += list(cons_3pplug-0.01)
df_dict['Algorithm'] += ['ConGDA']*10

df_dict['Constraint-Value'] += list(cons_el-0.01)
df_dict['Algorithm'] += ['ConElp']*10
```

```python
cons_df = pd.DataFrame(df_dict)
```

```python
fig = plt.figure(figsize =(6, 4)) 
sns.set(font_scale=2)
# Creating axes instance 
ax = fig.add_axes([0, 0, 1, 1]) 
  
graph = sns.barplot(x='Algorithm', y='Constraint-Value', data=cons_df, ax=ax, bottom=0)
# ax.fill_between([-1, 0, 1, 2, 3, 4, 6.5], -0.01, 0, facecolor="green", alpha=0.5, zorder=25)

# plt.xlabel(fontsize=20)
graph.axhline(0, linestyle='--', c='r', linewidth=2)
plt.ylim(-0.02)
plt.xlim(-0.5, 6.5)
plt.xticks(rotation=60)
plt.xlabel('')
plt.ylabel('Max Coverage Violation')
plt.savefig("./plots/" + name + "_" + split + "_cov_barplot.pdf", bbox_inches="tight")
plt.show() 
```
