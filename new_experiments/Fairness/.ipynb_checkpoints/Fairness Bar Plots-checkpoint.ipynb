```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
name = "default"
split = "test"
```

```python
obj_fw = np.load('./Results/fwunc_' + name + '_' + split + '_scores.npy')
obj_lr = np.load('./new-results/logreg_' + name + '_' + split + '_scores.npy')
obj_ballr = np.load('./new-results/ballogreg_' + name + '_' + split + '_scores.npy')
obj_sbfw = np.load('./Results/sbfw_' + name + '_' + split + '_scores.npy')
obj_3p = np.load('./Results/3player_' + name + '_' + split + '_scores.npy')
obj_el = np.load('./new-results/ellipsoid_' + name + '_' + split + '_scores.npy')
obj_3pplug = np.load('./new-results/3pplug_' + name + '_' + split + '_scores.npy')
```

```python
obj_3pplug
```

```python
df_dict = {
    'Algorithm': [],
    'Gmean': []
}

df_dict['Gmean'] += list(1-obj_lr)
df_dict['Algorithm'] += ['Plg (0-1)']*10

df_dict['Gmean'] += list(1-obj_ballr)
df_dict['Algorithm'] += ['Plg-Bal']*10

df_dict['Gmean'] += list(1-obj_fw)
df_dict['Algorithm'] += ['FW']*10

df_dict['Gmean'] += list(1-obj_3p)
df_dict['Algorithm'] += ['TFCO']*10

df_dict['Gmean'] += list(1-obj_sbfw)
df_dict['Algorithm'] += ['SplitFW']*10

df_dict['Gmean'] += list(1-obj_3pplug)
df_dict['Algorithm'] += ['GDA']*10

df_dict['Gmean'] += list(1-obj_el)
df_dict['Algorithm'] += ['Ellipsoid']*10
```

```python
scores_df = pd.DataFrame(df_dict)

fig = plt.figure(figsize =(6, 4)) 
sns.set(font_scale=2)
# Creating axes instance 
ax = fig.add_axes([0, 0, 1, 1]) 
sns.barplot(x = 'Algorithm', y = 'Gmean', data = scores_df)
# plt.xlabel(fontsize=20)
plt.ylim(min(df_dict['Gmean']) - 0.05)
plt.xlabel('')
plt.ylabel('G-mean Loss')
plt.xticks(rotation=60)
plt.savefig("./plots/" + name + "_" + split + "_gmean_barplot.pdf", bbox_inches="tight")
plt.show() 
```

```python
cons_fw = np.load('./Results/fwunc_' + name + '_' + split + '_cons.npy')
cons_ballr = np.load('./new-results/ballogreg_' + name + '_' + split + '_cons.npy')
cons_lr = np.load('./new-results/logreg_' + name + '_' + split + '_cons.npy')
cons_sbfw = np.load('./Results/sbfw_' + name + '_' + split + '_cons.npy')
cons_3p = np.load('./Results/3player_' + name + '_' + split + '_cons.npy')
cons_el = np.load('./new-results/ellipsoid_' + name + '_' + split + '_cons.npy')
cons_3pplug = np.load('./new-results/3pplug_' + name + '_' + split + '_cons.npy')
```

```python
df_dict = {
    'Algorithm': [],
    'Constraint-Value': []
}

df_dict['Constraint-Value'] += list(cons_ballr-0.05)
df_dict['Algorithm'] += ['Plg (0-1)']*10

df_dict['Constraint-Value'] += list(cons_ballr-0.05)
df_dict['Algorithm'] += ['Plg-Bal']*10

df_dict['Constraint-Value'] += list(cons_fw-0.05)
df_dict['Algorithm'] += ['FW']*10

df_dict['Constraint-Value'] += list(cons_3p-0.05)
df_dict['Algorithm'] += ['TFCO']*10

df_dict['Constraint-Value'] += list(cons_sbfw-0.05)
df_dict['Algorithm'] += ['SplitFW']*10

df_dict['Constraint-Value'] += list(cons_3pplug-0.05)
df_dict['Algorithm'] += ['GDA']*10

df_dict['Constraint-Value'] += list(cons_el-0.05)
df_dict['Algorithm'] += ['Ellipsoid']*10
```

```python
cons_df = pd.DataFrame(df_dict)
  
fig = plt.figure(figsize =(6, 4)) 
sns.set(font_scale=2)
# Creating axes instance 
ax = fig.add_axes([0, 0, 1, 1]) 
  
graph = sns.barplot(x='Algorithm', y='Constraint-Value', data=cons_df, ax=ax)
    
# ax.fill_between([-1, 0, 1, 2, 3, 4, 6.5], 0, 0.05, facecolor="green", alpha=0.5, zorder=25)
# plt.xlabel(fontsize=20)
graph.axhline(0, linestyle='--', c='r', linewidth=2)

# plt.ylim(0.0)
# plt.ylim(-0.05, 0.02)
plt.xlim(-0.5, 6.5)
plt.xlabel('')
plt.ylabel('Equal Opp. Violation')
plt.xticks(rotation=60)
plt.savefig("./plots/" + name + "_" + split + "_eopp_barplot.pdf", bbox_inches="tight")
plt.show() 
```
